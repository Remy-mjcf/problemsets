# [It's Always Sunny](https://www.practiceprobs.com/problemsets/regular-expressions-in-python/intermediate/its-always-sunny/)

Here's a quote from [It's Always Sunny in Philadelphia](https://en.wikipedia.org/wiki/It%27s_Always_Sunny_in_Philadelphia).

```python
import re
quote = (
    "That's TA, TR's ex-girlfriend - this is classic TA. TR broke up with TA because MK said that "
    "she saw TA flirting with WT at a party OK, but she was only doing it to make TR jealous because, you know, "
    "she thought that TR secretly liked EH, but TR didn't like EH. It was all a bunch of bull."
)
```

This quote has numerous abbreviated names consisting of two capital letters. For example, "TR" stands for "Trey".

Given the following dictionary of `abbreviation: full name` mappings,

```python
abbrevs = {
    "TA": "Tammy",
    "TR": "Trey",
    "MK": "Maureen Kinallen",
    "WT": "Walt Timby",
    "EH": "Erin Hannabry"
}
```

```python


def replace_abbreviations(input_string, abbreviation_dict):
    '''
    # Example usage
    abbreviations = {'FYI': 'For Your Information', 'ASAP': 'As Soon As Possible', 'BRB': 'Be Right Back'}
    input_text = "FYI, I'll be back ASAP. BRB!"

    result = replace_abbreviations(input_text, abbreviations)
    print(result)
    In this example, the replace_abbreviations function takes an input string and a dictionary of abbreviations and their meanings. It uses a regular expression to find word boundaries \b and matches any of the abbreviations. The sub method is then used to replace the matched abbreviations with their corresponding meanings.

    The output of the example usage would be:
    '''
    # Create a regular expression pattern to match abbreviations
    pattern = re.compile(r'\b(?:' + '|'.join(re.escape(key) for key in abbreviation_dict.keys()) + r')\b')
    
    # Replace abbreviations with their meanings
    result_string = pattern.sub(lambda x: abbreviation_dict[x.group()], input_string)
    
    return result_string
```

```python
new_str = replace_abbreviations(quote, abbrevs)
print(new_str)
```
